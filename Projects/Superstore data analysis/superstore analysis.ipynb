# Data visualization

## Plot a graph profit Vs quantity

```python
plt.bar(df['Quantity'],df['Profit'])
plt.title('Profit Vs Quantity',fontsize=15,color='red')
plt.xlabel('Quantity',fontsize=15)
plt.ylabel('Profit',fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.bar('Sub-Category','Category',data=df,color='green')
plt.show()
```

```python
print(df['State'].value_counts())
plt.figure(figsize=(10,4))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(df['Sub-Category'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(df['Quantity'],df['Sales'],color='red')
plt.title('Sales Vs Quantity',fontsize=15,color='blue')
plt.xlabel('Quantity',fontsize=15)
plt.ylabel('Sales',fontsize=15)
plt.grid(True)
plt.show()
```

```python
sns.relplot(x='Quantity',y='Profit',hue='Segment',data=df)
plt.show()
```

```python
sns.countplot(x='Segment',data=df,palette='coolwarm')
plt.show()
```

```python
plt.title('Segment',fontsize=20)
df['Segment'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
sns.countplot(x='Category',data=df,palette='tab10')
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
plt.title('Region',fontsize=20)
df['Region'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',palette='coolwarm_r')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=df,palette='coolwarm_r')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
df['Region'].value_counts()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='coolwarm_r')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['green','orange'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```
