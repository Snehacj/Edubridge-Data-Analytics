### count plot of  platforms

```python
plt.figure(figsize=(14,8))
sns.countplot(x=w['Platforms'])
plt.xlabel('Platforms',fontsize=15,color="Blue")
plt.ylabel('Count',fontsize=15,color="Blue")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

### Pie chart to find the count of platforms

```python
plt.title('Platforms',fontsize=20,color='Blue')
w['Platforms'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```
