```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.DataFrame({'product_name':productname,'price':product_price,'rating':rating})
df
```

```python
plt.figure(figsize=(18,10))
plt.bar('product_name','price',data=df,color='red')
plt.xticks(rotation=90)
plt.show()
```
