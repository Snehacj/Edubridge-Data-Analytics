## Machine Learning

### Feature Transforamtion

```python
t.dtypes
```

```python
t['OverTime']=t['OverTime'].replace({'Yes':1, 'No':0})
```

```python
t['Department']=t['Department'].replace({'Sales':0, 'Research & Development':2, 'Human Resources':1})
```

```python
t['MaritalStatus']=t['MaritalStatus'].replace({'Single':0, 'Married':1, 'Divorced':2})
```

```python
t['BusinessTravel']=t['BusinessTravel'].replace({'Travel_Rarely':1,'Travel_Frequently':2,'Non-Travel':0})
```

```python
t['Gender']=t['Gender'].replace({'Female':1,'Male':1})
```

```python
t.dtypes
```

```python
x=t.drop('JobSatisfaction',axis=1)
y=t.JobSatisfaction
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
```

```python
performance_report = classification_report(y_test,y_predict)
performance_report
```

```python
from sklearn.metrics import confusion_matrix
performance = confusion_matrix(y_test,y_predict)
performance
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model,x_test,y_test)
plt.show()
```
